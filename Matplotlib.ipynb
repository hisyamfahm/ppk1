Visualisasi Data dengan Matplotlib
============================
- Visualisasi data di Python dapat dilakukan dengan package <code>matplotlib</code>
- Pastikan package <code>matplotlib</code> sudah terinstall
- Kemudian import modul <code>pyplot</code> dalam package <code>matplotlib</code>
<pre><code>import matplotlib.pyplot as plt</code></pre>

```python
import matplotlib.pyplot as plt
```

### Menyiapkan data yang akan di plot
- Baca atau buat data yang akan di visualisasikan
- Data umumnya dibuat dengan tipe data koleksi maupun <code>numpy</code> array
- Jika data yang digunakan membutuhkan operasi numerik, disarankan mengguakan <code>numpy</code> array
- Data yang harus disiapkan adalah nilai pada sumbu x dan nilai pada sumbu y

```python
import numpy as np
x = np.linspace(0,10,100)
y = x**2
```

### Plot Data
- Lakukan plot grafik garis 2 dimensi sederhana menggunakan function <code>plot()</code>
- Function <code>plot()</code> bisa kita berikan 2 argumen yaitu, nilai pada sumbu x dan nilai pada sumbu y
- Terakhir, tampilkan grafik dengan function <code>show()</code> (Jika menggunakan jupyter, otomatis grafik akan ditampilkan tanpa memanggil function ini)

```python
plt.plot(x,y)
plt.show()
```

- Jika kita melakukan plot lebih dari satu kali, maka akan ditampilkan beberapa grafik dalam satu figure (gambar)

```python
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

### Menambahkan warna dan marker
- Untuk menambahkan warna dari grafik, bisa langsung kita definisikan dengan string pada parameter ketiga pada function <code>plot()</code>.
- Nilai warna yang diberikan dapat berupa 
    - singkatan warna ('b'), 
    - nama warnanya ('blue), 
    - hex string ('#0000ff'), 
    - RGB/RGBA tuple (0,0,1,1), atau 
    - intensitas grayscale dalam bentuk string ('0.8')  
- Contoh: 
<pre><code>plt.plot(x,y,'b')
plt.plot(x,y,'blue')
plt.plot(x,y,'#0000ff')
plt.plot(x,y,(0,0,1))</code></pre>
- Marker atau Line Style dapat didefinisikan dengan cara yang sama dengan warna di atas, tinggal menambahkan kode marker setelah kode warna
- Contoh: 
<pre><code>plt.plot(x,y,'b.') # plot titik warna biru tanpa garis
plt.plot(x,y,'b--') # plot garis putus-putus warna biru
plt.plot(x,y,'bo-') # plot garis biru dengan marker lingkaran</code></pre>
- Format string untuk color dan marker dapat mengacu pada <a href='https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html'>Matplotlib Documentation</a>
- Penulisan argument color dan marker juga dapat ditulis menggunakan keyword argument
- Contoh: 
<pre><code>plt.plot(x,y,color='blue',marker='.') # plot titik warna biru tanpa garis</code></pre>

```python
# Contoh plot grafik fungsi sin dan cos dengan color dan marker yang berbeda

x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y)

y2 = np.cos(x)
plt.plot(x,y2,'r')

y3 = np.sin(2*x)
plt.plot(x,y3,'go-')

y4 = np.cos(2*x)
plt.plot(x,y4,marker='x',color='orange')

plt.show()
```

### Menambahkan Legend
- Pada setiap melakukan plot grafik bisa kita tambahkan argument <code>label</code> untuk menjelaskan grafik tersebut
- Panggil function <code>legend()</code> untuk menambahkan legend dari grafik
- Posisi default dari legend yang ditambahkan adalah di kiri bawah dari grafik
- Jika kita ingin mengubah posisinya, tambahkan argument <code>loc</code> pada function <code>legend()</code>, yang nilainya adalah string posisi legend, misalnya 'bottom right', 'upper left', 'upper center', dll


```python
x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y, label=r'$ \mathcal{y = sin(x)} $')

y2 = np.cos(x)
plt.plot(x,y2,'r', label=r'$ \mathcal{y = cos(x)} $')

plt.legend(loc='upper center')
plt.show()
```

### Figure, Axes, Subplot
- Figure adalah objek window yang berisi semua objek untuk menggambar grafik, seperti sumbu, garis, teks, dan label
- Mendefinisikan objek figure dapat dilakukan dengan memanggil function <code>figure()</code>
- Axes adalah objek koordinat Cartesian untuk menggambar grafik (satu figure bisa lebih dari satu axes)
- Mendefinisikan objek axes dapat dilakukan dengan memanggil function <code>axes()</code>

```python
plt.figure()
ax1 = plt.axes()
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')

# Perintah di bawah akan menampilkan grafik pada figure yang berbeda
plt.figure()
plt.plot(x,2*y,'r')
plt.title('Grafik 2sin(x)')

plt.show()
```

```python
fig = plt.figure()
ax1 = plt.subplot(121) # plot 1 baris 2 kolom
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')
ax1.axis([0,6,-2,2])
ax1.set_xlabel('radian')
ax1.set_ylabel('sin(x)')

# Perintah di bawah ini akan menambahkan grafik di sebelah kanan grafik pertama
ax2 = plt.subplot(122)
ax2.plot(x,2*y,'r')
ax2.set_title('Grafik 2sin(x)')
ax2.axis([0,6,-2,2])
ax2.set_xlabel('radian')
ax2.set_ylabel('2sin(x)')
plt.show()
```
